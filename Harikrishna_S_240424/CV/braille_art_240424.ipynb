```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
pic=cv2.imread("./braille art1/roshumble.png")
plt.imshow(pic)
plt.title('Initial Image')
plt.axis('off')
plt.show()
```

```python
a=np.array(pic)
b=np.zeros((512,512,3))
gray = np.mean(a, axis=2).astype(np.uint8)
b = np.stack((gray,)*3, axis=-1)
plt.imshow(b)
plt.title('b/w Image')
plt.axis('off')
plt.show()
```

```python
kersharp=np.array([[-1,-1,-1],[-1,9,-1],[-1,-1,-1]])
sharp=cv2.filter2D(b,-1,kersharp)


plt.imshow(sharp)
plt.title('sharp Image')
plt.axis('off')
plt.show()
```

```python
edge=sharp-b
plt.imshow(edge)
plt.title('edge Image')
plt.axis('off')
plt.show()
edge=cv2.resize(edge, (128,128), interpolation=cv2.INTER_NEAREST)

```

```python
def br(dots):
    bits = sum(1 << (d - 1) for d in dots if 1 <= d <= 8)
    return chr(0x2800 + bits)


for i in range(128):
  for j in range(128):
    if edge[i][j][0]>0 :
      print(chr(0x2800),end='')
    else :
      print(br([1, 2, 3, 4, 5, 6, 7, 8]),end='')
  print()

```
