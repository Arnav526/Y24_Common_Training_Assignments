````python
import cv2
import numpy as np
from google.colab import files
from IPython.display import display, Markdown

def convert_to_braille(image_path, output_txt='braille_output.txt', output_width_chars=100):
    # Load and convert image to grayscale
    image = cv2.imread(image_path)
    if image is None:
        print("Error: Image not found.")
        return

    gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

    # Target size (based on Braille cell size)
    braille_cell_w = 2
    braille_cell_h = 4
    target_w_px = output_width_chars * braille_cell_w
    scale_ratio = target_w_px / gray.shape[1]
    target_h_px = int(gray.shape[0] * scale_ratio)
    target_w_px -= target_w_px % braille_cell_w
    target_h_px -= target_h_px % braille_cell_h

    # Resize and threshold the image
    resized = cv2.resize(gray, (target_w_px, target_h_px), interpolation=cv2.INTER_AREA)
    binary = cv2.adaptiveThreshold(resized, 255, cv2.ADAPTIVE_THRESH_GAUSSIAN_C,
                                   cv2.THRESH_BINARY_INV, 11, 2)

    # Braille dot mapping
    dot_map = {
        (0, 0): 0x01, (1, 0): 0x08,
        (0, 1): 0x02, (1, 1): 0x10,
        (0, 2): 0x04, (1, 2): 0x20,
        (0, 3): 0x40, (1, 3): 0x80,
    }

    braille_lines = []

    for y in range(0, target_h_px, 4):
        line = ''
        for x in range(0, target_w_px, 2):
            dots = 0
            for dy in range(4):
                for dx in range(2):
                    if binary[y + dy, x + dx] > 0:
                        dots |= dot_map[(dx, dy)]
            line += chr(0x2800 + dots)
        braille_lines.append(line)

    # Save to text file
    with open(output_txt, 'w', encoding='utf-8') as f:
        for line in braille_lines:
            f.write(line + '\n')

    # Print in notebook using Markdown (monospace)
    braille_str = '\n'.join(braille_lines)
    print("Braille Art:")
    display(Markdown(f"```\n{braille_str}\n```"))

    print(f"Braille Art saved to {output_txt}")

# Upload image
uploaded = files.upload()

# Process uploaded image(s)
for filename in uploaded:
    convert_to_braille(filename)

# Auto-download
files.download('braille_output.txt')

````
