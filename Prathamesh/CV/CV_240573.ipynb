```python
import cv2
import numpy as np
from google.colab.patches import cv2_imshow
from google.colab import files

files.upload()
img = cv2.imread("Tralalero-Tralala.png")
img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
cv2_imshow(img)
```

```python
h, w = img.shape

ph = (4 - h % 4) % 4
pw = (2 - w % 2) % 2

img = np.pad(img, ((0, ph), (0, pw)), mode='constant', constant_values=0)

_, bin_img = cv2.threshold(img, 100, 255, cv2.THRESH_BINARY)
cv2_imshow(bin_img)
```

```python
out = []

for y in range(0, bin_img.shape[0], 4):
    row = ''
    for x in range(0, bin_img.shape[1], 2):
        blk = bin_img[y:y+4, x:x+2]
        code = 0
        bits = [
            (0, 0, 0), (1, 0, 1), (2, 0, 2),
            (0, 1, 3), (1, 1, 4), (2, 1, 5),
            (3, 0, 6), (3, 1, 7),
        ]
        for r, c, b in bits:
            if blk[r, c]:
                code |= 1 << b
        row += chr(0x2800 + code)
    out.append(row)

print('\n'.join(out))
```
